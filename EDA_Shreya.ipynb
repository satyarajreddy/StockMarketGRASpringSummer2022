```python
import numpy as np
import pandas as pd
# import pandas_datareader as pdr
import matplotlib.pyplot as plt
```

### KOSS

```python
df_koss = pd.read_csv("KOSS.csv")
df_koss.head()
```

```python
df_koss.info()
```

```python
# convert to pandas datetime
df_koss['Date'] = pd.to_datetime(df_koss['timestamp'])
df_koss.head()
```

```python
df_koss.info()
```

```python
# using the quarter property
df_koss['Quarter'] = df_koss['Date'].dt.quarter

# display the dataframe
df_koss.head()
```

```python
# using the to_period() function
df_koss['Quarter_Info'] = df_koss['Date'].dt.to_period('Q')

# display the dataframe
df_koss.head()
```

```python
# using the to_period() function
df_koss['Yearly_Info'] = df_koss['Date'].dt.to_period('Y')

# display the dataframe
df_koss.head()
```

```python
# using the to_period() function
df_koss['Monthly_Info'] = df_koss['Date'].dt.to_period('M')

# display the dataframe
df_koss.head()
```

```python
df_koss.to_csv('KOSS_Updated.csv')
```

## SIMPLE MOVING AVERAGE

```python
# Simple Moving Average
# We would need to reduce the sample size to get a better reading, we can take a yearly data set and then do a SMA with 20 day rate

# def get_sma(prices, rate):
#     return prices.rolling(rate).mean()

df_koss.index = np.arange(df_koss.shape[0]) # Convert the index to array from [0, 1, 2, ...number of rows]
closing_prices = df_koss['close'] # Use only closing prices

# sma = get_sma(closing_prices, 20) # Get 30 day SMA
sma = closing_prices.rolling(20).mean()
df_koss.head()

# Plot the data
plt.title('KOSS' + ' SMA')
plt.xlabel('Days')
plt.ylabel('Closing Prices')
plt.plot(closing_prices, label='Closing Prices')
plt.plot(sma, label='20 Day SMA')
plt.legend()
plt.show()
```

## BOLLINGER BAND

```python
# Bollinger Band - Bollinger bands are calculated two standard deviations away from the SMA
# We would need to reduce the sample size to get a better reading, we can take a yearly data set and then do a SMA and STD Deviation with 20 day rate

def get_bollinger_bands(prices, rate=500):
#     sma = get_sma(prices, rate) # <-- Get SMA for 20 days
    sma = closing_prices.rolling(20).mean()
    std = prices.rolling(rate).std() # <-- Get rolling standard deviation for 20 days
    bollinger_up = sma + std * 2 # Calculate top band
    bollinger_down = sma - std * 2 # Calculate bottom band
    return bollinger_up, bollinger_down

df_koss.index = np.arange(df_koss.shape[0])
closing_prices = df_koss['close']

bollinger_up, bollinger_down = get_bollinger_bands(closing_prices)

plt.title('KOSS' + ' Bollinger Bands')
plt.xlabel('Days')
plt.ylabel('Closing Prices')
plt.plot(closing_prices, label='Closing Prices')
plt.plot(bollinger_up, label='Bollinger Up', c='g')
plt.plot(bollinger_down, label='Bollinger Down', c='r')
plt.legend()
plt.show()
```
