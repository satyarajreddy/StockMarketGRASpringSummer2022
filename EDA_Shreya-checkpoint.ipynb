## Exponential Moving Average

```python
# Common function for finding the EMA based on the formula
# EMA is a type of moving average indicator that gives greater weight or importance to previous stock prices. 
# The essential difference between EMA and SMA is that EMA responds faster to upward price movement compared to SMA

def calculate_ema(prices, days, smoothing=2):
    ema = [sum(prices[:days]) / days]
    for price in prices[days:]:
        ema.append((price * (smoothing / (1 + days))) + ema[-1] * (1 - (smoothing / (1 + days))))
    return ema

```

## KOSS Exponential Moving Average

```python
# EMA invocation and plotting with a 20 day range of observations

koss_ema = calculate_ema(df_koss['close'], 20)

price_X = np.arange(df_koss.shape[0])
ema_X = np.arange(20, df_koss.shape[0]+1)
 
# Comparison plot b/w stock values & EMA
plt.title('KOSS' + ' EMA')
plt.xlabel('Days')
plt.ylabel('Price')
plt.plot(price_X, df_koss['close'], label='Closing Prices')
plt.plot(ema_X, koss_ema, label='EMA')
plt.legend()
plt.show()
```

## KOSS Comparison plot b/w stock values, SMA & EMA

## BOLLINGER BAND Function

```python
# Bollinger Band - Bollinger bands are calculated two standard deviations away from the SMA
# We would need to reduce the sample size to get a better reading, we can take a yearly data set and then do a SMA and STD Deviation with 20 day rate
# Defining a common function to be used across all the symbols

def get_bollinger_bands(prices, rate=500):
#     sma = get_sma(prices, rate) # <-- Get SMA for 20 days
    sma = prices.rolling(20).mean()
    std = prices.rolling(rate).std() # <-- Get rolling standard deviation for 20 days
    bollinger_up = sma + std * 2 # Calculate top band
    bollinger_down = sma - std * 2 # Calculate bottom band
    return bollinger_up, bollinger_down

```

## KOSS BOLLINGER BAND

```python
df_koss.index = np.arange(df_koss.shape[0])
closing_prices = df_koss['close']

koss_bollinger_up, koss_bollinger_down = get_bollinger_bands(koss_closing_prices)

plt.title('KOSS' + ' Bollinger Bands')
plt.xlabel('Days')
plt.ylabel('Closing Prices')
# plt.plot(closing_prices, label='Closing Prices')
koss_closing_prices.plot(label='Closing Prices', figsize=(16,5))
plt.plot(koss_bollinger_up, label='Bollinger Up', c='g')
plt.plot(koss_bollinger_down, label='Bollinger Down', c='r')
plt.legend()
plt.show()
```
